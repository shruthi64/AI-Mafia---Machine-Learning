```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
test=pd.read_csv(r"Test.csv")
train=pd.read_csv(r"Train.csv")
```

```python
print(train.head())
print(train.shape,test.shape,sep="\n")
```

```python
x1=train['feature_1'].values.reshape((-1,1))
x2=train['feature_2'].values.reshape((-1,1))
x3=train['feature_3'].values.reshape((-1,1))
x4=train['feature_4'].values.reshape((-1,1))
x5=train['feature_5'].values.reshape((-1,1))
x=train[['feature_1','feature_2','feature_3','feature_4','feature_5']].values
y=train['target'].values.reshape((-1,1))
print(x,y,x.ndim,y.ndim,sep="\n\n")
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x1,y,color="green",label='feature_1')
plt.scatter(x2,y,color="cyan",label='feature_2')
plt.scatter(x3,y,color="orange",label='feature_3')
plt.scatter(x4,y,color="blue",label='feature_4')
plt.scatter(x5,y,color="magenta",label='feature_5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
lr=LinearRegression()
lr.fit(x,y)
```

```python
output=lr.predict(x)
output
```

```python
bias=lr.intercept_
coef=lr.coef_
print(bias,coef)
```

```python
lr.score(x,y)
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,output,label='predictons')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
t1=test['feature_1'].values.reshape((-1,1))
t2=test['feature_2'].values.reshape((-1,1))
t3=test['feature_3'].values.reshape((-1,1))
t4=test['feature_4'].values.reshape((-1,1))
t5=test['feature_5'].values.reshape((-1,1))
t=test[['feature_1','feature_2','feature_3','feature_4','feature_5']].values
print(t,t.shape,type(t),sep="\n\n")
```

```python
test_output=lr.predict(t)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(t1,test_output,color="green",label='feature_1')
plt.scatter(t2,test_output,color="cyan",label='feature_2')
plt.scatter(t3,test_output,color="orange",label='feature_3')
plt.scatter(t4,test_output,color="blue",label='feature_4')
plt.scatter(t5,test_output,color="magenta",label='feature_5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
plt.close()
```

```python
plt.figure(figsize=(10,8))
plt.plot(t,test_output,label='predictons')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
